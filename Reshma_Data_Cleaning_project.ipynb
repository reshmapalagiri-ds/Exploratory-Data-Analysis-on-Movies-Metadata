```python
rows = []  #for all movies
genres = list(uniqueGenres)
for i in data['genres']:
  genList = ast.literal_eval(i)
  movieGenres= [x['name'] for x in genList]
  row = [] #this row is to store the 1s and o's
  for j in genres:
    row.append(int(j in movieGenres))
  rows.append(row)
```

#Data Visualisation

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Profit Distribution
sns.histplot(dataBudgetRevenue['profit'],bins = 30,kde = True)
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
#Budget vs Revenue
sns.scatterplot(data = dataBudgetRevenue,x = 'budget',y = 'revenue')
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.show()
```

```python
#Boxplot of profit by genre
sns.boxplot(data = dataBudgetRevenue,x = 'Action',y = 'profit')
plt.title('Profit by Action Genre')
plt.xlabel('Genre')
plt.ylabel('Profit')
plt.show()
```

```python
#Profir over time
dataBudgetRevenue['release_date'] = pd.to_datetime(dataBudgetRevenue['release_date'],errors = 'coerce')
dataBudgetRevenue['release_year'] = dataBudgetRevenue['release_date'].dt.year
sns.lineplot(x='release_year',y = 'profit',data = dataBudgetRevenue)
plt.title('Profit over Time')
plt.xlabel('Year')
plt.ylabel('Profit')
plt.show()
```

```python
#Correlaton Heatmap
corr_matrix = dataBudgetRevenue[['budget','revenue','profit']].corr()
sns.heatmap(corr_matrix,annot = True,cmap = 'coolwarm')  #to visualise correlation between numerical features
plt.title('Correlation Heatmap')
plt.show()
```
